```python
#Initial imports
import numpy as np
import pandas as pd
import hvplot.pandas
import plotly.express as px
```

```python
# Load Google Trends model data
trends_df = pd.read_csv('../resources/trends.csv', index_col=0, infer_datetime_format=True, parse_dates=True)
trends_df = trends_df.drop(columns=['Real'])
trends_df = trends_df.rename(columns={'Predicted':'Google Trends Model Predicted Price'})
trends_df
```

```python
# Load Fear and Greed Index model data
fng_df = pd.read_csv('../resources/fng_model.csv', index_col=0, infer_datetime_format=True, parse_dates=True)
fng_df = fng_df.drop(columns=['Real'])
fng_df = fng_df.rename(columns={'Predicted':'Fear and Greed Index Model Predicted Price'})
fng_df
```

```python
# Load Multivariate model data
multi_df = pd.read_csv('../resources/multi.csv', index_col=0, infer_datetime_format=True, parse_dates=True)
multi_df = multi_df.drop(columns=['Real'])
multi_df = multi_df.rename(columns={'Predicted':'Multivariate Model Predicted Price'})
multi_df
```

```python
# Load Bitcoin price data
btc_price_df = pd.read_csv('../resources/BTC-USD.csv', index_col=0, infer_datetime_format=True, parse_dates=True)
btc_price_df = btc_price_df.drop(columns = ['Open', 'High', 'Low', 'Adj Close', 'Volume'])
btc_price_df = btc_price_df.rename(columns={'Close':'Bitcoin Price'})
btc_price_df
```

```python
# Join the data into a single DataFrame
join_df = pd.concat([trends_df, fng_df, multi_df, btc_price_df], axis=1, join="inner")
join_df
```

```python
# Plot
ax = join_df['Bitcoin Price'].plot(figsize=(20,10))
join_df['Multivariate Model Predicted Price'].plot(ax=ax)
join_df['Fear and Greed Index Model Predicted Price'].plot(ax=ax)
join_df['Google Trends Model Predicted Price'].plot(ax=ax)

#Legend
ax.legend(['Bitcoin Price', 
           'Multivariate Model Predicted Price', 
           'Fear and Greed Index Model Predicted Price', 
           'Google Trends Model Predicted Price'])
```
